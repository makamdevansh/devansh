#Code snippet to display which country was the largest oil producing country from the year range of 1932 to 2014

```python
for year in range(1932,2015):

    city=data[data["year"]==year]["cty_name"]
    oil_pro=data[data["year"]==year]["oil_prod32_14"]

    my_dict = dict(zip(oil_pro,city))
    max_pro=oil_pro.max()
    max_city=my_dict[max_pro]
    print("In year",year,max_city,"has produced maximum oil of",max_pro,"gallons")
```

#Code snippet to display a dictionary with keys as the top 4 oil producing countries and values as the no.of times they were at the top

```python
final_dict={}
for year in range(1932,2014):

    city=data[data["year"]==year]["cty_name"]
    oil_pro=data[data["year"]==year]["oil_prod32_14"]

    my_dict = dict(zip(oil_pro,city))
    max_pro=oil_pro.max()
    max_city=my_dict[max_pro]

    if max_city in final_dict.keys():
        final_dict[max_city]+=1
    else:
        final_dict[max_city]=1

print(final_dict)
```

#Code snippet to display a pie chart with top 4 oil producing countries vs number f times it produced Using mattplotlib  

```python
y=final_dict.values()
x=final_dict.keys()

plt.pie(y, labels=x,startangle=90,shadow=True,explode=(0,0,0,0,0.2),autopct='%1.1f%%' )
plt.title("piechart of city vs no of times it produced maximum oil")
plt.show()
```

#code snippet to estimate the production of oil in up coming years using LinerRegression.

   The linear regression model provides a sloped straight line representing the relationship between the variables. Used to data prediction.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
data=data.replace(float('NaN'),0)
city=input('enter the city which you want to check : ')
# Input data: previous years and respective oil produced
previous_years = data[data["cty_name"].str.contains(city,case=False)]["year"].tolist() # Replace with actual years
oil_produced = data[data["cty_name"].str.contains(city,case=False)]["oil_prod32_14"].tolist()  # Replace with actual oil production

# Convert lists to numpy arrays
X = np.array(previous_years).reshape(-1, 1)
y = np.array(oil_produced)

# Create a linear regression model
model = LinearRegression()
model.fit(X, y)

# Predict oil production for upcoming years
upcoming_year = range(2015,2051)  # Replace with the years you want to predict
upcoming_year = np.array(upcoming_year).reshape(-1, 1)

predicted_production = model.predict(upcoming_year)

# Plot the data and regression line
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, model.predict(X), color='red', label='Regression Line')
plt.scatter(upcoming_year, predicted_production, color='green', label='Predicted Data')
plt.xlabel('Year')
plt.ylabel('Oil Production')
plt.legend()
plt.title("oil production in {}".format(city))
plt.show()
print("city")
# Print the predicted production for upcoming years
for year, production in zip(upcoming_year, predicted_production):
    print(f"Year: {int(year)}, Predicted Production: {production:.2f} gallons")
```
